# 3. 힙(Heap)으로 생일 저장 및 출력

````python
import csv
from heap import Heap


def make_date_key(row):
    try:
        year  = int(row["생년"])
        month = int(row["생월"])
        day   = int(row["생일"])
        if year < 1900 or year > 2100:
            raise ValueError(f"비정상 연도: {year}")
        return (year, month, day)
    except (ValueError, KeyError):
        return None


people = []
with open("birthday.csv", encoding="utf-8") as f:
    for row in csv.DictReader(f):
        key = make_date_key(row)
        if key is None:
            print(f"[예외처리] {row['이름']}: 생년월일 데이터 오류 → 힙에서 제외")
        else:
            people.append((key, row["이름"]))

heap = Heap()
for key, name in people:
    heap.insert((key, name))

print("\n[생년월일 순서 TOP 10 (생일 느린 순)]")
print(f"{'순위':<5} {'이름':<8} 생년월일")
print("-" * 28)
rank = 1
while not heap.isEmpty() and rank <= 10:
    key, name = heap.deleteMax()
    year, month, day = key
    print(f"{rank:<5} {name:<8} {year}년 {month:>2}월 {day:>2}일")
    rank += 1
````

# 4. 원형 이중 연결 리스트로 우리 조 생년월일 출력

````python
import csv
from circularDoublyLinkedList import CircularDoublyLinkedList


# 우리 조 명단
MY_TEAM = {"이채윤", "구수진", "권지영", "김수민", "김주영",
           "김태연", "김효린", "서지윤", "유지민", "이승주"}


def format_birthday(row):
    try:
        year  = int(row["생년"])
        month = int(row["생월"])
        day   = int(row["생일"])
        if year < 1900 or year > 2100:
            raise ValueError(f"비정상 연도: {year}")
        return f"{year}년 {month}월 {day}일"
    except (ValueError, KeyError):
        return "생년월일 정보 없음"


cdll = CircularDoublyLinkedList()
with open("birthday.csv", encoding="utf-8") as f:
    for row in csv.DictReader(f):
        if row["이름"] in MY_TEAM:
            cdll.append(dict(row))

print("[우리 조 생년월일 목록]")
print(f"{'이름':<8} 생년월일")
print("-" * 25)
for person in cdll:
    name     = person["이름"]
    birthday = format_birthday(person)
    print(f"{name:<8} {birthday}")
````

## Q01
**최대 힙에서 더 얕은 곳의 원소가 더 깊은 곳의 원소보다 더 작은 값을 가질 수 있는가?**

**있다.**  
최대 힙 조건은 '부모 ≥ 자식'이며, 서로 다른 서브트리 간에는 크기 순서가 보장되지 않는다.  
예를 들어 루트의 왼쪽 자식(깊이 1)이 오른쪽 서브트리의 손자(깊이 2)보다 작을 수 있다.

```
        10
       /  \
      3    9
            \
             8
```
깊이 1의 노드 3 < 깊이 2의 노드 8 → 가능

## Q02
**최대 힙 A[0…n-1]에서 마지막 원소 A[n-1]은 항상 가장 작은 값을 갖는가?**

**아니다.**  
마지막 원소는 힙의 마지막 레벨 중 가장 오른쪽 리프일 뿐이다.  
같은 레벨에 있는 다른 리프가 더 작은 값을 가질 수 있으므로 A[n-1]이 반드시 최솟값은 아니다.

```
        10
       /  \
      7    8
     / \
    3   6
```
A[n-1] = 6이지만 최솟값은 3이다.

## Q03
**buildHeap()에서 스며내리기를 하지 않고 그냥 넘어가는 노드 수는 정확히 몇 개인가?**

**⌊n/2⌋ 개**  
buildHeap()은 리프 노드에는 스며내리기를 수행하지 않는다.  
n개의 노드 중 리프 노드 수는 정확히 ⌊n/2⌋이므로, 그냥 넘어가는 노드는 **⌊n/2⌋**개이다.

## Q04
**스며내리기(sift-down) 알고리즘의 최악/최선 시간은? (Θ-표기)**

- **최악의 경우 : Θ(log n)**  
  루트에서 리프까지 내려가야 하는 경우로, 트리 높이 ⌊log n⌋번 비교·교환이 발생한다.

- **최선의 경우 : Θ(1)**  
  첫 번째 비교에서 부모가 두 자식보다 이미 크거나 같아 바로 종료된다.

---

## Q05
**힙의 맨 마지막 원소를 삭제하는 작업은 간단한가?**

**간단하다.**  
마지막 원소 A[n-1]은 리프 노드이므로 배열 크기를 1 줄이는 것(n = n-1)으로 충분하다.  
힙 조건을 깨뜨리지 않으므로 스며내리기나 스며오르기가 전혀 필요 없다. **O(1)** 연산이다.

---

## Q06
**buildHeap()을 '위쪽에서 스며오르기 반복'으로 구현하면 더 효율적인가?**

**아니다. 더 비효율적이다.**

| 방법 | 점근적 시간 |
|------|------------|
| 교재 방식 (아래→위, 스며내리기) | **Θ(n)** |
| 위→아래, 스며오르기 반복 | **Θ(n log n)** |

스며오르기는 아래쪽 노드(수가 많음)일수록 이동 거리가 길어져 총 비용이 Θ(n log n)이 된다.  
반면 스며내리기 방식은 아래쪽에 노드가 많더라도 이동 거리가 짧아 합산하면 Θ(n)이 된다.

## Q07
**최대 힙에서 임의의 원소 값이 증가했을 때 O(log n)으로 힙을 수선하는 방법은?**

값이 증가한 경우 해당 노드가 부모보다 커질 수 있으므로 **스며오르기(sift-up)** 를 수행한다.

**과정:**
1. 값이 증가한 노드의 인덱스를 i로 설정한다.
2. A[i] > A[⌊(i-1)/2⌋] (부모) 이면 두 값을 교환한다.
3. i를 부모 인덱스로 갱신하고, 힙 조건을 만족할 때까지 2를 반복한다.
4. 루트에 도달하거나 부모 ≥ 자식이 되면 종료한다.

트리 높이가 O(log n)이므로 전체 시간복잡도는 **O(log n)** 이다.
