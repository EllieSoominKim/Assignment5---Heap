```python
import csv
from heap import Heap


def make_date_key(row):
    try:
        year  = int(row["생년"])
        month = int(row["생월"])
        day   = int(row["생일"])
        if year < 1900 or year > 2100:
            raise ValueError(f"비정상 연도: {year}")
        return (year, month, day)
    except (ValueError, KeyError):
        return None


people = []
with open("birthday.csv", encoding="utf-8") as f:
    for row in csv.DictReader(f):
        key = make_date_key(row)
        if key is None:
            print(f"[예외처리] {row['이름']}: 생년월일 데이터 오류 → 힙에서 제외")
        else:
            people.append((key, row["이름"]))

heap = Heap()
for key, name in people:
    heap.insert((key, name))

print("\n[생년월일 순서 TOP 10 (생일 느린 순)]")
print(f"{'순위':<5} {'이름':<8} 생년월일")
print("-" * 28)
rank = 1
while not heap.isEmpty() and rank <= 10:
    key, name = heap.deleteMax()
    year, month, day = key
    print(f"{rank:<5} {name:<8} {year}년 {month:>2}월 {day:>2}일")
    rank += 1
```

```python
import csv
from circularDoublyLinkedList import CircularDoublyLinkedList


# 우리 조 명단
MY_TEAM = {"이채윤", "구수진", "권지영", "김수민", "김주영",
           "김태연", "김효린", "서지윤", "유지민", "이승주"}


def format_birthday(row):
    try:
        year  = int(row["생년"])
        month = int(row["생월"])
        day   = int(row["생일"])
        if year < 1900 or year > 2100:
            raise ValueError(f"비정상 연도: {year}")
        return f"{year}년 {month}월 {day}일"
    except (ValueError, KeyError):
        return "생년월일 정보 없음"


cdll = CircularDoublyLinkedList()
with open("birthday.csv", encoding="utf-8") as f:
    for row in csv.DictReader(f):
        if row["이름"] in MY_TEAM:   # 우리 조만 추가
            cdll.append(dict(row))

print("[우리 조 생년월일 목록]")
print(f"{'이름':<8} 생년월일")
print("-" * 25)
for person in cdll:
    name     = person["이름"]
    birthday = format_birthday(person)
    print(f"{name:<8} {birthday}")
```
